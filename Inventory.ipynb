# Product Details

1. Product ID
2. Name of Product
3. Price of Product
4. Quantity

```python
# Noting the time of trasaction
import time

# Reading the Inventory
fd=open("Inventory.txt",'r')
products=fd.read().split("\n")
fd.close()

# Taking User Input
ui_username        = input("Enter your Name: ")
ui_phone_number    = input("Enter your Phone No: ")
ui_mail_id         = input("Enter your Mail: ")
ui_product_id      = input("Enter the product id:")
ui_product_quantity= input("Enter the product quantity:")

updated_prod_list=[]

# Going through each product detail
for product in products:
    product_details=product.split(",")
    
    # Checking if product exists or not
    if(product_details[0]==ui_product_id):
       
        # If we're having enough quantity
        if(int(ui_product_quantity)<=int(product_details[3])):
            
            print(".................................")
            print("Product Name     :",product_details[1])
            print("Product Price    : $",product_details[2])
            print("Product Quantity :",ui_product_quantity)
            print(".................................")
            print("Billing Amount   : $",int(product_details[2])*int(ui_product_quantity))
            print(".................................")
            
            # Updating Inventory list
            product_details[3] = str(int(product_details[3]) - int(ui_product_quantity))
            
            # Generating Sales in Sales.txt
            fd = open("Sales.txt",'a')
            sales_detail = ui_username +","+ ui_phone_number +","+ ui_mail_id +","+product_details[1] +","+ ui_product_id +","+ ui_product_quantity +","+ str(int(ui_product_quantity) * int(product_details[2]))+","+time.ctime()+ "\n"
            fd.write(sales_detail)
            fd.close()
            
        else:
       # If we don't have enough quantity
            print("Sorry we don't have enough quantity")
            print("We only have",product_details[3],"quantity")
            print("Would you like purchase it?")
            
            ch=input("Press Y/N: ")
           
            if(ch=='Y'or ch=='y'):
            # If you want to purchase with remaining quantity
            
                print(".................................")
                print("Product Name     :",product_details[1])
                print("Product Price    : $",product_details[2])
                print("Product Quantity :",product_details[3])
                print(".................................")
                print("Billing Amount   : $",int(product_details[2])*int(product_details[3]))
                print(".................................")
                
                # Generating Sales in Sales.txt
                fd = open("Sales.txt",'a')
                sales_detail = ui_username +","+ ui_phone_number +","+ ui_mail_id +","+product_details[1] +","+ ui_product_id +","+ product_details[3] +","+ str(int(product_details[3]) * int(product_details[2]))+","+time.ctime()+ "\n"
                fd.write(sales_detail)
                fd.close()
                
                # Updating Inventory list
                product_details[3]='0'
            
            else:
                print("Thanks for shopping!")
                
    # Updating my Inventory List        
    updated_prod_list.append(product_details)
       
    

lst = []

# Updating my Inventory String
for i in updated_prod_list:
    prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
    lst.append(prod)
    
# Removing Last \n from the list
lst[-1] = lst[-1][:-1]

# Updating Inventory File    
fd = open('Inventory.txt','w')
for i in lst:
    fd.write(i)

fd.close()

print("-------------------")
print("Inventory Updated")
```
